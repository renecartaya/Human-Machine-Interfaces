```python
###############################################################################################
#
#                            Real Time Plot Graphical User Interface
#                                  
################################################################################################

# Works in connection with the AnalogReadSerial_firm_II

from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure
import tkinter as tk
import numpy as np
import serial as ser

# ========================= Main windown settings ============================= #

root = tk.Tk()
root.title('Real Time Plot')
root.configure(background = '#e0e0e0')
root.iconbitmap('C:/Users/renea/Pictures/Logos/gen_logo.ico')
root.geometry("520x500") # set the window size

# ====================== Serial port configuration =========================== #

serial_port = ser.Serial('COM5',9600);
serial_port.reset_input_buffer()

# ======================= global variables =========================== #

data = np.array([])
cond = False

# ========================= Functions Set ============================= #

# =========================== plot data =============================== #

def plot_data():
    global cond, data
    
    if (cond == True):
        
        a = serial_port.readline()
        a.decode()
        
        if(len(data) < 100):
            data = np.append(data,float(a[0:4]))
        else:
            data[0:99] = data[1:100]
            data[99] = float(a[0:4])

        lines.set_xdata(np.arange(0,len(data)))
        lines.set_ydata(data)
        
        canvas.draw()
    
    root.after(1,plot_data)

# =========================== conditionals ============================= #
    
def plot_start():
    global cond
    cond = True
    serial_port.reset_input_buffer()

def plot_stop():
    global cond
    cond = False
    
def plot_reset():
    global cond
    cond = False

def test():
    global cond
    cond = False
    
# ================== create Plot object on GUI ============================ #
# add figure canvas

fig = Figure();
ax = fig.add_subplot(111)

#ax = plt.axes(xlim=(0,100),ylim=(0, 120)); #displaying only 100 samples
ax.set_title('Serial Data input');
ax.set_xlabel('Independent Variable')
ax.set_ylabel('Imput Signal (V)')
ax.set_xlim(0,100)
ax.set_ylim(-0.5,6)
lines = ax.plot([],[])[0]

canvas = FigureCanvasTkAgg(fig, master=root)  # A tk.DrawingArea.
canvas.get_tk_widget().place(x = 10,y=10, width = 500,height = 400)
canvas.draw()

# ================================= buttons set ============================================= #

# ============================= Original Winglets ============================================ #
root.update();
start = tk.Button(root, text = "Start", font = ('calbiri',12),command = lambda: plot_start())
start.place(x = 170, y = 450 )

root.update();
stop = tk.Button(root, text = "Stop", font = ('calbiri',12), command = lambda:plot_stop())
stop.place(x = start.winfo_x()+start.winfo_reqwidth() + 20, y = 450)

root.update();
reset = tk.Button(root, text = "Reset", font = ('calbiri',12), command = lambda:plot_stop())
reset.place(x = stop.winfo_x()+stop.winfo_reqwidth() + 20, y = 450)

# ============================= test Winglets ============================================ #

root.update();
test = tk.Button(root, text = "test", font = ('calbiri',12), command = lambda:test())
test.place(x = stop.winfo_x()+stop.winfo_reqwidth() + 70, y = 550)


# =============================== Clossing ========================================== #

root.after(1,plot_data)
root.mainloop()
serial_port.close()  
```
